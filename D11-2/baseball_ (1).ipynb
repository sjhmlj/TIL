```python
import requests
from bs4 import BeautifulSoup

naver_baseball= "https://sports.news.naver.com/kbaseball/record/index?category=kbo"
kbo_rank= requests.get(naver_baseball)
kbo_rank_list= BeautifulSoup(kbo_rank.content, "html", from_encoding='utf-8')
```

```python
# team_rank_list= kbo_rank_list.select('.tbl_box')  
# team_rank_list
```

```python
team_rank_list= kbo_rank_list.select('#regularTeamRecordList_table > tr')
team_rank_list
```

```python
team_rank_list
```

```python
for top in team_rank_list:
    rank= top.select('#regularTeamRecordList_table > tr > th > strong')
    name= top.select('#regularTeamRecordList_table > tr > td.tm > div')
    game= top.select('#regularTeamRecordList_table > tr > td:nth-child(3) > span')
    win= top.select('#regularTeamRecordList_table > tr > td:nth-child(7) > strong')
    on_base= top.select('#regularTeamRecordList_table > tr > td:nth-child(10) > span')


    #rank= top.select('th > strong')
    #name= top.select('div.span')      
    print(rank, name, game, win, on_base)
```
